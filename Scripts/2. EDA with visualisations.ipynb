```python
#1. Correlation heatmap
df_patent.corr()
```

```python
# Create a correlation heatmap using matplotlib

plt.matshow(df_patent.corr())
plt.show()
```

```python
# Add labels, a legend, and change the size of the heatmap

f = plt.figure(figsize=(20, 20)) # figure size 
plt.matshow(df_patent.corr(), fignum=f.number) # type of plot
plt.xticks(range(df_patent.shape[1]), df_patent.columns, fontsize=18, rotation=90) # x axis labels
plt.yticks(range(df_patent.shape[1]), df_patent.columns, fontsize=18) # y axis labels
cb = plt.colorbar() # add a colour legend (called colorbar)
cb.ax.tick_params(labelsize=14) # add font size
plt.title('Correlation Matrix', fontsize=30) # add title
```

3. Categorical Plots

```python
#allowing full output to display
pd.set_option('display.max_rows', None)
```

```python
#creating a categorical varibale from the 'Total Patent Numbers' column
#checking the frequency of this column
df_patent['Total patent numbers'].value_counts()
```

```python
#calculating the quartiles for the patent numbers column
df_patent['Total patent numbers'].quantile(0.25)
```

```python
df_patent['Total patent numbers'].quantile(0.5)
```

```python
df_patent['Total patent numbers'].quantile(0.75)
```

```python
#creating categories from the 'Total patent numbers' column
df_patent.loc[df_patent['Total patent numbers'] < 2.0, 'Patent count category'] = 'Low'
```

```python
df_patent.loc[(df_patent['Total patent numbers'] >= 2.0) & (df_patent['Total patent numbers'] < 5.0), 'Patent count category'] = 'Middle'
```

```python
df_patent.loc[df_patent['Total patent numbers'] >= 24.0, 'Patent count category'] = 'High'
```

```python
#checking new 'Patent count category' column
df_patent['Patent count category'].value_counts()
```

```python
#creating the catplot in seaborn using the patent count categories above
sns.set(style="ticks")
g = sns.catplot(x="GDP PPP (per capita)", y="Total patent numbers", hue="Patent count category", data=df_patent)
```

```python
#this chart gives us a little more information that the previous visualisation. From this we can see that while the middle
#and low patent categories are clustered throughout the x axis, with no discernable pattern, the 'high category' has some more interesting
#patterns that might be worth exloring
```

```python
sns.set(style="ticks")
g = sns.catplot(x="Ferrtility Rate (15-19 yo)", y="Total patent numbers", hue="Patent count category", data=df_patent)
```

```python
#same as above - while low and middle patent category countries are spread across the x axis, countries that produce 
#high number of patents are correlated with lower adolecsent fertility rates
```

```python
sns.set(style="ticks")
g = sns.catplot(x="R&D (% GDP)", y="Total patent numbers", hue="Patent count category", data=df_patent)
```

```python
#same as above, with the extra catgroisation of patent counts we can see that while spending on R&D doesn't necessarily
#produce higher patent counts, countries with higher patent counts tend to spend more on R & D.
```
